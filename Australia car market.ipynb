```python
df_carsinfo['Year'] = pd.to_numeric(df_carsinfo['Year'])
```

```python
df_carsinfo['Car Age'] = np.where(df_carsinfo['Year'] > 2000, 'Modern Classic Car', 'Classic Car')

print(df_carsinfo)
```

```python
df_carsinfo['Car Age'].value_counts()
```

```python
df_carsinfo['Mileage'] = np.where(df_carsinfo['Kilometers'] > 100000, 'High Mileage', 'Low mileage')

print(df_carsinfo)
```

```python
df_carsinfo['Mileage'].value_counts()
```

```python
df_carsinfo['Price Range'] = np.where(df_carsinfo['Price'] > 100000, 'Expensive', 'Affordable')

print(df_carsinfo)
```

```python
df_carsinfo['Price Range'].value_counts()
```

```python
df_pivot = df_carsinfo.pivot_table(values ='Brand', columns='Price Range', aggfunc='count')
```

```python
df_pivot
```

```python
df_pivot.plot(kind='bar', xlabel='Brand', ylabel='Price Range', title='Classification by Price Range', figsize=(10,8))
```
